```python
class bst:
    def __init__(self, elements):
        elements = np.array(elements)
        median = np.median(elements)
        self.root = bst.node(value = int(median), depth = 0)

        for val in elements:
            self.insert(val)

    def insert(self, val):
        if self.root is not None:
            self.root.insert(val)
        else:
            self.root = bst.node(value = val, depth = 0)
    
    def find(self, val):
        if self.root is None:
            return False
        else:
            return self.root.find(val)
    
    def remove(self, val):
        if self.root is None:
            return False
        else:
            if self.root == val:
                self.root = None
            if not(self.find(val)):
                return
            else:
                self.root.remove(val)
    
    def visualize(self):
        return self.root.visualize()
        

    class node:
        def __init__(self, value = None, parent = None, left = None, right = None, depth = None):
            self.value = value
            self.parent = parent
            self.left = left
            self.right = right
            self.depth = depth
        
        def insert(self, value, curr_depth=0):
            print(f"Depth: {curr_depth}, Node value: {self.value}")
            if self.value == value:
                return
            elif value > self.value:
                if self.right is None:
                    self.right = bst.node(value=value, parent=self, depth=curr_depth + 1)
                else:
                    self.right.insert(value, curr_depth = (curr_depth + 1))
            else:
                if self.left is None:
                    self.left = bst.node(value=value, parent=self, depth=curr_depth + 1)
                else:
                    self.left.insert(value, curr_depth = (curr_depth + 1))
        
        def __repr__(self):
            return f"Value = {self.value}, Depth = {self.depth}"
        
        def find(self, val):
            if self.value == val:
                return True
            else:
                if self.left is not None:
                    self.left.find(val)
                if self.right is not None:
                    self.right.find(val)
            return False

        def remove(self, val):
            if self.value == val:
                if self.right is None and self.left is None:
                    parent = self.parent
                    if parent.left is self:
                        parent.left = None
                    else:
                        parent.right = None
                    return
                elif self.right is not None and self.left is not None:
                    if self.right.value > self.left.value:
                        
                    else:
                        pass


            else:
                if self.left is not None:
                    self.left.remove(val)
                if self.right is not None:
                    self.right.remove(val)
                return

        
        def visualize(self):
            print(self)
            if self.left is not None:
                print("Left")
                self.left.visualize()
            if self.right is not None:
                print("Right")
                self.right.visualize()
            return False
```
