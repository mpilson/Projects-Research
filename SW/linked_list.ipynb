```python
class linked_list:
    def __init__(self, array):
        if len(array) == 0:
            self.head = None
        else:
            self.head = self.node(array[0])
            prev = self.head
            for i in range(1,len(array)):
                curr = self.node(array[i], left = prev)
                prev.right = curr
                prev = curr

    def __repr__(self):
        s = ""
        curr = self.head
        while curr is not None:
            if curr == self.head:
                s += f"{curr.value} "
            else:
                s += f"--> {curr.value}"
            curr = curr.right
        return s
    
    def append(self,value):
        if self.head is not None:
            curr = self.head
            while curr.right is not None:
                curr = curr.right
            new = self.node(value, left = curr)
            curr.right = new
        else:
            self.head = self.node(value)
    
    def length(self):
        count = 0
        curr = self.head
        while curr is not None:
            curr = curr.right
            count += 1
        return count
    
    def find(self,value):
        curr = self.head
        while curr.right is not None:
            if curr.value == value:
                return value
            else:
                curr = curr.right
        return None
            

    def pop(self):
        if self.head is None:
            raise IndexError("pop from empty list")
        
        curr = self.head
        if curr.right is None:
            value = self.head.value
            self.head = None
            return value

        while curr.right.right is not None:
            curr = curr.right
        
        value = curr.right.value
        curr.right = None
        return value
    
    def prepend(self,value):
        if self.head is not None:
            old = self.head
            new = self.node(value, right = old)
            self.head = new
            old.left = new
        else:
            self.head = self.node(value)

    def delete(self,value):
        if self.head is not None:
            curr = self.head
            while curr.right is not None:
                if curr.value == value:
                    left = curr.left
                    right = curr.right
                    if left is not None:
                        left.right = right
                    if right is not None:
                        right.left = left
                    if curr == self.head:
                        self.head = right
                    return curr.value
                else:
                    curr = curr.right
            else:
                raise IndexError("delete value not present")
        else:
            raise IndexError("delete from empty list")
    
    class node:
        def __init__(self, value, left = None, right = None):
            self.value = value
            self.left = left
            self.right = right
```
